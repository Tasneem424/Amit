```python
#Task_1
email= "Amit_ml@gmail.edu"

result="valid email" if email.count('@') == 1 and '.' in email.split('@')[1] else "invalid email"
print(result)

usr_name=email.split('@')[0]
print(f"User name : {usr_name}")

x=email.split('@')[1]
domain=x.split('.')[0]
print(f"Domain : {domain}")

dom_type=x.split('.')[1]
if (dom_type=='com'):
     print("Commercial Domain")
elif (dom_type=='edu'):
     print("Educational Domain")
else :
     print("Other Domain")
```

```python
##Task_2
message = "###!!@mocleW EPGTQ!!!6789"

main_message="".join(c for c in message if c.isalpha() or c.isspace())
main_message
x=main_message.split(" ")
x
x1=x[0]
x1=x1[::-1]
x1
x2=x[1].replace('E',' ')
x2
final_de_message=x1+x2
print(f"Final Decoded Message : {final_de_message}")
```

```python
##Task_3
Encoded_Message="&&&**$gnirtS PLIO!!@1234"
main_message="".join(c for c in Encoded_Message if c.isalpha() or c.isspace())
main_message=main_message.split(" ")
x1=main_message[0][::-1]
x2=main_message[1].replace('I','E').replace('O','U')
x2
final_de_message=x1+" "+x2
print(f"Final Decoded Message : {final_de_message}")

```
