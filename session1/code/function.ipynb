```python
def user_info(name, age=30):
    """
    this function takes name and age from user and returns a formatted string   
    param_1:takes the user name from input user interface
    type param_1 : str
     param_2:takes the user age from input user interface 
     type param_2 : int
     this function will print the name and age of user in a formatted string
     return : formatted string with name and age
     type str,int
    """
    return f"{name} is {age} years"
```

```python
user_info( age=30, name="Alice")
```

```python
user_info("bob")
```

```python
def user_in(*params):
    """
    this function takes variable number of parameters and returns them as a tuple
    param_1:takes variable number of parameters from input user interface
    type param_1 : tuple
    this function will return the parameters as a tuple
    return : tuple of parameters
    type : tuple
    """
    print(params)
    print(type(params))
```

```python
user_in('apple', 'banana', 'cherry',20,17,"ppp")
```
