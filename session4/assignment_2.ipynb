```python
import os
import random
class thanos_project():
     def __init__(self,fld_path,num_files):
          self.fld_path=fld_path
          self.num_files=num_files

     def open_fld(self):
          if  not os.path.exists(self.fld_path):
             os.mkdir(self.fld_path)
          else:
             print("directory already exists",self.fld_path)

     def create_fl(self):
         for i in range(self.num_files):
             inner_path=os.path.join(self.fld_path,f"file_{i}.txt")
             if not os.path.exists(inner_path):
                 os.mkdir(inner_path)

     def count_fl(self):
         return len(os.listdir(self.fld_path))
     
     def del_half(self):
         files = os.listdir(self.fld_path)
         half = len(files) // 2
         files_to_delete = random.sample(files, half)

         for file in files_to_delete:
             os.remove(os.path.join(self.fld_path, file))

     def check(self):
         return self.count_fl()
     

t=thanos_project("C:\\Users\\tasne\\New folder (2)\\Amit\\session4\\assignmnent_2_folder",10)
t.open_fld()
t.create_fl()
print(t.count_fl())
t.del_half()
print(t.check())
     
```
