```python
#Q1
class F():
     """
     class handles the operation of getting 
     fibonacchi sereie for the given number
     """
     def __init__(self,num):
         self.num=num

     def fib(self,num):
       """
       get a list of fibonacchi series for a given number
       return:
            a list of fibonacchi series for a given number
       """
       if num==1:
           return [0]
       elif num==2:
           return [0,1]
       else :
            fibonacchi=self.fib(num-1)
            fibonacchi.append(fibonacchi[-1]+fibonacchi[-2])
       return fibonacchi
     
c=F()
c.fib(10)
       
```

```python
#Q2
class my_set():
     """
     class handles the process of removing a certain item from a list
     """
     def __init__(self,myset,num):
          self.myset=myset
          self.num=num
     def remove_item(self):
          """
          this method remove a certain item for a given set
          return:
              the edited set
          """
          self.myset.discard(self.num)
          return self.myset
     
S=my_set({1,2,3,4,5},3)
S.remove_item()

```

```python
#Q3
class my_set():
     """
     class handles the process of removing mutiple items from a given set
     """
     def __init__(self,myset,num):
          self.myset=myset
          self.num=num
     def remove_items(self):
          """
          this method remove mutiple items from a given set
          return:
              the edited set
          """
          for item in self.num:
             self.myset.discard(item)
          return self.myset
     
S=my_set({1,2,3,4,5},[1,4])
S.remove_items()
```

```python
#Q4
class tuple_op():
     """
     this class get the index of certain item in tuple
     """
     def __init__(self,my_tuple,item):
          self.my_tuple=my_tuple
          self.item=item
     def get_index(self):
          """
          this method get the index of certain item in tuple
          return:
              the index of the certain item
          """
          return self.my_tuple.index(self.item)
     
D=tuple_op(('apple','banana','orange','grape','banana'),'orange')
D.get_index()
```

```python
#Q5
class splitter():
     """
     this class split the scentence into words
     """
     def __init__(self,scentence):
          self.scentence=scentence
     
     def split_scen(self):
          """
          this method split the scentence into words
          return:
              splitting words
          """
          return self.scentence.split()
     
S=splitter("welcome to amit")
S.split_scen()
```

```python
#Q6
class check():
     """
     this class check palindrome
     """
     def __init__(self,word):
          self.word=word

     def check_palindrome(self):
          """
          this method check if the word is palindrome
          return:
             "palindrome" if the word is palindrome
          """
          rev=self.word[::-1]
          if rev== self.word:
               return "palindrome"
          else :
               return "not palindrome"
          
C=check("racecar")
C.check_palindrome()
```

```python
#Q7
class Fact():
     """
     this class get factorial of a number
     """
     def __init__(self,num):
          self.num=num
     def get_fact(self):
          """
          this method get factorial of a number
          return:
             factorial of a number
          """
          fact=1
          for i in range(1,self.num+1):
             fact=fact*i
          return fact

f=Fact(5)
f.get_fact()
         
```

```python
#Q8
class squre_num():
     """this class get the square of a given numbers"""
     def __init__(self,num1,num2):
          self.num1=num1
          self.num2=num2
     def get_sqr(self):
          """
          get the squuare of a given numbers
          return :
              the square of numbers
          """
          list=[]
          for i in range(self.num1,self.num2+1):
              list.append(i*i)

          return list
     
Sqr=squre_num(1,10)
Sqr.get_sqr()
```

```python
#Q9
class sum_ov():
     """get the sum of even numbers"""
     def __init__(self,lst):
         self.lst=lst
     def sum_even(self):
         """
         this method get the sum of even number in list
         return:
              sum of even number
         """
         sum=0
         for i in range (len(self.lst)):
             if self.lst[i] % 2==0:
                sum+=self.lst[i]
         return sum 
ev=sum_ov([1,2,3,4,5,6])
ev.sum_even()
```
