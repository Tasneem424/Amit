```python
def factorial(num):
     if num==0:
          return 1
     if num==1:
          return 1
     return num*factorial(num-1)
```

```python
def is_prime(num):
     for i in range(2,num):
          if num%i==0:
               return False
     return True
```

```python
is_prime(11)
```

```python
def reverse_txt(text):
    split_txt=text.split()
    new_list=split_txt[::-1]
    result=" ".join(new_list)
    return result
```

```python
reverse_txt("welcome depi")
```
